```python
df = pd.DataFrame(data)
df.info()
```

```python
#transform non-numeric labels to numeric labels

from sklearn.preprocessing import LabelEncoder  # label encoder used to normalize labels
l = LabelEncoder()
for i in df.columns:
  if df[i].dtype == 'object':        #transform non-numerical labels (as long as they are hashable and comparable) to numerical labels
    df[i]=l.fit_transform(df[i])
```

```python
df.shape
```

**Splitting data**


```python
x = df.drop('mental_fitness', axis=1)
y = df['mental_fitness']

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=.20,random_state=2)
```

```python
print("xtrain: ",xtrain.shape)
print("xtest: ",xtest.shape)
print("\nytrain: ",ytrain.shape)
print("ytest: ",ytest.shape)
```

**Applying regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
lr.fit(xtrain,ytrain)

y_predict = lr.predict(xtrain)

# mean_squared_error
mse = mean_squared_error(ytrain, y_predict)

# root mean square error
rmse = (np.sqrt(mse))

# coefficient of determination
r2 = r2_score(ytrain, y_predict)

print("PERFORMANCE OF THE REGRESSION MODEL\n")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))


```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain,ytrain)

y_pred = rf.predict(xtrain)

mse = mean_squared_error(ytrain, y_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(ytrain, y_pred)

print("PERFORMANCE OF THE RANDOM FOREST REGRESSION MODEL\n")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))
```

**EVALUATION** **(Comparing the performances of both Linear Regression and Random Forest Regressor models)**

```python
# Linear regression

y_predict = lr.predict(xtrain)
mse = mean_squared_error(ytrain, y_predict)
rmse = (np.sqrt(mse))
r2 = r2_score(ytrain, y_predict)

print("PERFORMANCE OF THE LINEAR REGRESSION MODEL\n")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))


# Random Forest Regressor

y_pred = rf.predict(xtrain)
mse = mean_squared_error(ytrain, y_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(ytrain, y_pred)

print("\n\nPERFORMANCE OF THE RANDOM FOREST REGRESSION MODEL\n")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 score is {}".format(r2))

```
